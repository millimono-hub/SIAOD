```python
for _ in range(int(input())):
    n, k, z = map(int, input().split())
    a = [int(x) for x in input().split()]
    ans = 0
    s = 0
    mx = 0
    for i in range(k + 1):
        if i < n - 1:
             mx = max(mx, a[i] + a[i + 1])
        s += a[i]
        if i % 2 == k % 2:
            tmp = (k - i) // 2
            if tmp <= z:
                ans = max(ans, s + mx * tmp)
print(ans)
```

–í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ –º–æ–∂–Ω–æ –ø–µ—Ä–µ–≤–µ—Ä–Ω—É—Ç—å –ø–æ–¥–º–∞—Å—Å–∏–≤ [3,5]. –¢–æ–≥–¥–∞ ùëé=[1,8,3,6,7,6] –∏ 1‚ãÖ5+8‚ãÖ9+3‚ãÖ6+6‚ãÖ8+7‚ãÖ8+6‚ãÖ6=235.
